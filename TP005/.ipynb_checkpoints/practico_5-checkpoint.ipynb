```python
MAT_RGB2YIQ = np.array([[0.299, 0.587, 0.114],
                        [0.596,-0.275,-0.321],
                        [0.211,-0.523, 0.311]])

def apply_matrix(img, M):
    return np.matmul(img.reshape((-1,3)), M.T).reshape(img.shape)

def rgb2yiq(img):
    return apply_matrix(img, MAT_RGB2YIQ)

def yiq2rgb(img):
    return apply_matrix(img, np.linalg.inv(MAT_RGB2YIQ))

def rmse(img1, img2):
    return np.sqrt(np.mean((img1-img2)**2))
```

```python
from matplotlib import cm

def plot_kernel(data, ax=None):
    rows, cols = data.shape
    y, x = np.meshgrid(np.arange(rows),np.arange(cols),indexing='ij')
    if ax == None:
        fig = plt.figure()
        ax = fig.gca(projection='3d')
    _min, _max = (np.min(data), np.max(data))
    ax.plot_surface(x, y, data.T, cmap=cm.jet, vmin=_min, vmax=_max)

def plot_images_and_kernel(img, img_filt, kernel):
    fig = plt.figure(figsize=(17,5))
    ax1 = fig.add_subplot(131)
    ax1.imshow(img, 'gray')
    ax1.title.set_text('Input image')
    ax2 = fig.add_subplot(132)
    ax2.imshow(img_filt, 'gray')
    ax2.title.set_text('Filtered image')
    ax3 = fig.add_subplot(133, projection='3d')
    plot_kernel(kernel, ax=ax3)
    ax3.title.set_text('Kernel')
    plt.show()
```

```python
# Ejemplo del kernel identidad

def kerID(tam):# kernel identidad
    kernel = np.zeros((tam,tam))
    kernel[np.floor(tam/2).astype(int), np.floor(tam/2).astype(int)] = 1
    return kernel
kernel = kerID(5)
img_filt = convolve2d(img, kernel, 'same')

plot_images_and_kernel(img, img_filt, kernel)
```

# Kernels pasa-bajo

```python
# Kernel cuadrado
def kerCuad(tam):
    kernel = np.ones((tam, tam))
    kernel = kernel / kernel.sum()
    return kernel

tam = 9 
kernel = kerCuad(tam)
img_filt = convolve2d(img, kernel, 'same')
plot_images_and_kernel(img, img_filt, kernel )
```

```python
# Kernel circular
def kerCirc(tam):
    tam = np.floor(tam / 2)
    X,Y = np.mgrid[-tam:tam+1:1, -tam:tam+1:1]
    RR = X*X + Y*Y
    kernel = np.where(RR <= tam*tam, 1,0)
    kernel = kernel / kernel.sum()
    return kernel

tam = 5
kernel = kerCirc(tam)
img_filt = convolve2d(img, kernel, 'same')
plot_images_and_kernel(img, img_filt, kernel )
```

```python
# Kernel Bartlett
def kerBartlett(tam):
    tam = np.floor(tam / 2)
    X,Y = np.mgrid[-tam:tam+1:1, -tam:tam+1:1]
    X = tam - np.abs(X) + 1
    Y = tam - np.abs(Y) + 1
    kernel = np.matmul(X, Y)/(2*tam + 1)
    kernel /= kernel.sum()
    return kernel

tam = 5
kernel = kerBartlett(tam)
img_filt = convolve2d(img, kernel, 'same')
plot_images_and_kernel(img, img_filt, kernel )
```

```python
# Kernel Gausiano

def kerGaus(tam, std):
    mu, sigma = 0, std # media y desvio estandar
    normal = stats.norm(mu, sigma)
    tam = np.floor(tam / 2)
    X,Y = np.mgrid[-tam:tam+1:1, -tam:tam+1:1]
    X = normal.pdf(X)
    Y = normal.pdf(Y)
    kernel = np.matmul(X,Y)
    kernel = kernel / kernel.sum()
    return kernel

tam = 11
kernel = kerGaus(tam, 0.5)
img_filt = convolve2d(img, kernel , 'same')

plot_images_and_kernel(img, img_filt, kernel )
```

# Kernels pasa-alto

```python
#  Kernel Laplace
def kerLaplace(opcion):# opcion 1 -> 4 vecinos; opcion 2 -> 8 vecinos
    if opcion == 1:
        kernel = np.array([[0, -1, 0],[-1, 4, -1],[0, -1, 0]])
    else:
        kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
    return kernel

kernel = kerLaplace(2) # opcion 1
img_filt = convolve2d(img, kernel, 'same')

plot_images_and_kernel(img, img_filt, kernel)
```

```python
# Kernel PA = Id - PA

def kerPA(kerPB): # Param: Kernel pasabajo
    N, M = kerPB.shape
    kerID = np.zeros( (N, M) )
    kerID[np.floor(N/2).astype(int), np.floor(N/2).astype(int)] = 1
    return kerID - kerPB

kernel = kerPA( kerCuad(13) )
img_filt = convolve2d(img, kernel, 'same')
plot_images_and_kernel(img, img_filt, kernel)
```

```python
# Kernel pasaBanda
def dog(kerGauss1, kerGauss2):# Param: dos kernel's gausianos
    return kerGauss1 - kerGauss2

kernel = dog(kerGaus(tam, 1), kerGaus(tam, 1.2))


img_filt = convolve2d(img, kernel, 'same')
plot_images_and_kernel(img, img_filt, kernel) 
```

```python
# Mejora de constraste

def mejContraste(kerPB):# param kernel pasaalto
    N, M = kerPB.shape
    kerID = np.zeros( (N, M) )
    kerID[np.floor(N/2).astype(int), np.floor(N/2).astype(int)] = 1
    return kerID + 0.2*kerPA( kerPB )

kernel = mejContraste( kerGaus(5, 0.2))


img_filt = convolve2d(img, kernel, 'same')
plot_images_and_kernel(img, img_filt, kernel) 
```

# Implementar los siguiente filtros direccionales (Asimétricos)

```python
# Sobel
#def filtSobel():
#    return

kernel = np.array([[-1, 0, 1],[-2, 0, 2],[-1, 0, 1]])
img_filt = convolve2d(img, kernel, 'same')
plot_images_and_kernel(img, img_filt, kernel) 

kernel = np.rot90(kernel)
img_filt = convolve2d(img, kernel, 'same')
plot_images_and_kernel(img, img_filt, kernel)

```

```python
# Sobel
# def filtSobel():
#    return

kernel = np.array([[-1, 0, 1],[-2, 0, 2],[-1, 0, 1]])

# Sobel horizontal
img_filt_H = convolve2d(img, kernel, 'same')
plot_images_and_kernel(img, img_filt_H, kernel) 

# Sobel vertical
img_filt_V = convolve2d(img, np.rot90(kernel) , 'same')
plot_images_and_kernel(img, img_filt_V, kernel)

# Detector de Bordes con filtro sobel
img_filt = np.sqrt(  img_filt_H * img_filt_H + img_filt_V * img_filt_V) 
img_filt2 = np.where(img_filt > 0.05, 1, 0)

plot_images_and_kernel(img, img_filt2, kernel)
```

Proceso
El proceso del algoritmo de detección de bordes de Canny se puede dividir en cinco pasos diferentes:

1.- Aplicar filtro gaussiano para suavizar la imagen con el fin de eliminar el ruido
2.- Encuentra los gradientes de intensidad de la imagen.
3.- Aplicar umbral de magnitud de gradiente o supresión de corte de límite inferior para deshacerse de la respuesta falsa a la detección de bordes
4.- Aplicar doble umbral para determinar posibles bordes
5.- Seguimiento de borde por histéresis: finalice la detección de bordes suprimiendo todos los demás bordes que son débiles y no están conectados a bordes fuertes.
